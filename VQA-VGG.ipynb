## VGG16

```python
import numpy as np
import pandas as pd
from keras.applications.vgg16 import VGG16, preprocess_input
from keras.utils import load_img, img_to_array
```

```python
df_train = pd.read_csv('df_train.csv')
```

```python
df_train.columns
```

```python
df_train.head()
```

```python
vgg16 = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
vgg16.summary()
```

```python
def extract_features(imagePath):
  img = load_img(imagePath, target_size=(224, 224))
  img = img_to_array(img)
  img = np.expand_dims(img, axis=0)
  img = preprocess_input(img)
  features = vgg16.predict(img)
  return features
```

```python
image_paths = df_train[4500:7000]['image_path']
features_array = np.array([])
for x in range(2500):
    features_array = np.append(features_array, extract_features(image_paths[x+4500]))
    print(x)
```

```python
features_array = features_array.reshape(2500,7,7,512)
```

```python
np.save('vgg_extracted_features_2500.npy',features_array)
```
